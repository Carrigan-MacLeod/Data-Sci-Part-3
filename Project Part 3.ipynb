# Introduction

This paper uses a species of fish, the threespine stickleback (*Gasterosteus aculeatus*), to observe how parental behavior changes based on environmental changes. Most significantly, this paper aims to investigate how changes in parental care behavior and differences in male phenotypes affect offspring survival in response to environmental change.

# Part 2

Chosen Article: Good parenting may not increase reproductive success under environmental extremes (Fox et al. 2018, Journal of Evolutionary Biology)



**Section 1**

In this experiment, pairs of sticklebacks were mated, and specific data were collected on the weight and size of the egg clutches laid by the females and the nests built by the males. Then, as embryos developed, the males and their eggs were placed in either high or low dissolved oxygen (DO) conditions, and the males' parental behavior, such as fanning and tending to the eggs, was observed in response to the change in conditions.

There are 19 total features on this data set collected from this experiment. They include:  

	The ID number of the male fish being studied

	The level of dissolved oxygen conditions under which the male was tested 

	The length of the male in millimeters 

	The mass of the mating female prior to spawning 

	Mass of the female after spawning 

	The spawned egg mass 

	Egg mass after 7 days post incubation [dpi] 

	Number of surviving eggs 7 dpi 

	Number of surviving eggs 7 dpi adjusted for weight 

	Scale out of 1-5, indicating the degree of embryo development based on pigmentation 

	Measure of total nest area taken 4 days after spawning 

	Measure of bulk nest area taken 4 days after spawning 

	Compactness of nest 

	Rate of nest gluing behavior 4 days after spawning 

	Rate of male tending during trial 

	Rate of male fanning during trial 

	Proportion of time spent tending during trial 

	Proportion of time spent fanning during trial 

	Whether the male and female within the trial laid eggs 


There are 61 observations, each representing an individual mating trial run on an individual male stickleback. Together, these features give us details on the relationship between DO levels and changes in parental care, as well as embryo health, in male threespine stickleback. 

With this experiment, the authors article aimed to answer the question of how the parental care behaviors of threespine stickleback change in response to environmental extremes, like high or low DO conditions, and how the size of the stickleback might influence the level of parental care. 

They hypothesized that in low DO conditions, male stickleback will express higher levels of fanning behavior, to provide more oxygen to their developing offspring and increase offspring development and survival, than males in high DO conditions. They also hypothesized that larger males would spend more time expressing fanning behavior than smaller males, due to their easy access to resources, whereas smaller males must spend more time and effort gathering resources. 

Another research question that could be answered using this data could be whether nest size impacts egg survival rate.

Next, import pandas and use the "read_csv" function to open the file in Jupyter. I recommend adding "index_col = 0" to this command as well in order to set the observations ("MaleID") as the index, as shoen below. 


```python
import pandas as pd

fish_data = pd.read_csv("DATA_Fox+et+al_Environmentally-dependent+parental+care.csv", index_col = 0)
fish_data.head(10)
```

From here, I began cleaning my data. Some of the code I used for this data frame are things we haven't learned in class, but are instead things I had to look up in order to clean up the data set nicely. 

Looking at the data frame generated from importing the CSV file, there were multiple things I wanted to change. This includes the excessive columns beginning with "Unamed..." due to trailing commas from the CSV file, rows with cells that read N/A, and rows where no offspring were produced. 

In order to clean the data, I chose to first focus on removing the unnamed columns. I looked up ways to do this, which was easy, as trailing commas are common with CSV files. I found that I could reassign fish_data to itself, but only keep all rows and **only** those columns whose names do not start with Unnamed.

```python
clean_col_data = fish_data.loc[:, ~fish_data.columns.str.contains("^Unnamed")]
```

Next, I wanted to remove any rows that had empty cells, which came out as N/A in this data frame. I did so by using a function we haven't learned in class, but is commonly used, called "dropna". This allowed me to automatically remove any rows that had N/A cells.

```python
clean_NA_data = clean_col_data.dropna()
```

Finally, I wanted to remove the last few rows left where males were unable to reproduce with the females presented to them. I did this by using a simple boolean that said to only keep rows where the value of Filter(noeggs) was **not** equal to 0. 

```python
cleaned_data = clean_NA_data[fish_data["Filter(noeggs)"] != 0]
```

After these steps, I received a cleaned, shortened data frame that was easy to read and use for further data analysis. 

```python
cleaned_data.head(10)
```

For my second visual, I wanted to create a visualization comparing the male body length (mm) to the proportion of time males spent tending to their nests in high and low DO conditions, something that was not represented in the visualizations included in this paper. This should, in theory, show the relation between the size of males and the time they spend tending to their offspring based on changes in the environment. This would contribute to the researcher's understanding of male size to parental care behavior, and maybe provide more insight into how this is impacted by environment.

I think this information would be best visualized using a histogram since it will show the distribution of individual fish among the range of sizes, clearly showing if there are any trends towards one end of the spectrum or the other. Additionally, it will show if any outliers are present, such as a few very large males that could drive the mean higher. 

```python
#Here I use a linear model plot to show the data as a scatter plot with regression lines ot show the trends of the data
sns.lmplot(
    data=cleaned_data, #Use my cleaned data set
    x = 'MaleLn_in', #Place male body size on the x-axis
    y = 'propten4', #Place proportion of tending behavior on y-axis
    hue = 'DO_Level') #Split the data into low and high DO conditions
```

This visual shows that the proportion of time spent tending to their nests actually decreased with increasing male size in low DO conditions, and oppositely increased in high DO conditions. However, this could be due to the fact that there is a very wide range in tending behavior among the 42 males in my cleaned data set. Additionally, when looking at the data points alone, without the regression line, there does not seem to be any real pattern, possibly meaning that there is no real correlation between the two features. This matches the results of the paper, which stated that through a statistical analysis, tending behavior was not impacted by male size, which refutes the hypothesis of Fox et al. that larger males would have a higher likelihood of changing their parental care behaviors in response to their environment.
