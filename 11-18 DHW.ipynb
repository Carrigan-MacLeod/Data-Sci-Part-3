1) One main purpose of GitHub is to easily share code and allow for simple, live collaboration between users.

   
2) Version control allows changes made to code or files to be recorded so specific versions can be accessed later in time. This can help me specifically by allowing me to see how my files have changed as I accept or work through edits I receive from the peer review process.


3) In a group project, GitHub can help manage collaboration by allowing for the review of versions and individual changes that are proposed before they are integrated into the code. This also allows for discussion and troubleshooting between groupmates.


4) An employer may want to look at a potential candidate's GitHub because it essentially acts as a portfolio of their work with coding and data management/organization. This will give them an idea of the candidate's skill set and give the employer a better idea of whether they want to hire the candidate.
